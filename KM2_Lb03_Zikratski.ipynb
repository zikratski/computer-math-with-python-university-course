```python
y = lambda x: (x-5)**3 #лямбда-фция - задает аргументы(x) и закон(x-5)**3
```

```python
x_min = 0.; x_max = 10.; step = 0.1
```

```python
x_array = np.arange(x_min, x_max, step) #создает массив x-ов
y_array = y(x_array) #создаем массив значений функции по x-ам
```

```python
matrix = np.transpose([x_array, y_array]) #создаем матрицу(транспонир.) с аргументами и значениямми функции
```

```python
fig1 = plt.figure()

y_min, y_max = np.min(y_array), np.max(y_array) #мин и макс значения функции

ax1 = plt.axes() #создание графической области
plt.axis([x_min-1, x_max+1, y_min-1, y_max+1]) #задание пределов по осям для нее

line1, = ax1.plot([],[],'b') #граф.объект line1 синего цвета
```

```python
line1, = ax1.plot([],[],'b') #граф.объект line1 синего цвета
print(type(line1))
line1.get_xdata()
```

```python
line1.get_ydata()
```

```python
def at_frame1(t): #будет вызываться в каждом кадре анимации
    """добавляет к объекту line1 точку с координатами (t[0],t[1])
    
    Arguments :
    
    t : массив из двух элементов
    
    Returns : None
    """
    x_coord = list(line1.get_xdata())
    y_coord = list(line1.get_ydata())
    
    x_coord.append(t[0])
    y_coord.append(t[1])
    
    line1.set_data(x_coord, y_coord)    
    return None

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
```

```python
def func(lambd):
    phi = lambd / 4
    x = np.sin(lambd) * np.cos(phi)
    y = np.cos(lambd) * np.cos(phi)
    z = np.sin(phi)
    return x, y, z
```

```python
def at_frame(t):
    ax.cla()  #каждый раз очищаем область видения
    ax.plot_wireframe(xs, ys, zs, color='gray', alpha=0.3) #рисуем сферу(проволочная)
    x, y, z = func(lambd[-t:])
    #ax.text(0,0,f'{t}')
    ax.plot(x, y, z, 'b')
    
```

```python
lambd = np.linspace(0, 2*np.pi, 100)
x, y, z = func(lambd[-10:])
z
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

lambd = np.linspace(0, 2*np.pi, 100)

u = np.linspace(0, 2 * np.pi, 100) #долгота
v = np.linspace(0, np.pi, 100) #широта
xs = np.outer(np.cos(u), np.sin(v))
ys = np.outer(np.sin(u), np.sin(v))
zs = np.outer(np.ones(np.size(u)), np.cos(v))

def init():
    ax.plot_wireframe(xs, ys, zs, color='gray', alpha=0.3)
    x, y, z = func(lambd)
    ax.plot(x, y, z, 'b')

FuncAnimation(fig, at_frame, frames=len(lambd), init_func=init, repeat=False, interval=10)
```
