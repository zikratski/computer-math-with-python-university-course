Срез X[i:j] означает "предоставить из последовательности X все содержимое, начиная с позиции i и заканчивая позицией j, не включая содержимое для позиции j".


```python
str1, str1[1:3], str1[2:-1]
```



Срез возвращает новый объект последовательности

### 2.8 Операции и методы для встроенных типов объектов



На объекте в Python выполняются ТОЛЬКО допустимые для его типа операции и методы. Это называется строгой типизацией объектов.

Множество операций и методов для конкретного типа объектов называется интерфейсом объекта.

После того, как объект создан, он навсегда привязывается к своему интерфейсу.

В языке Python определено фиксированное и неизменяемое отображение каждой операции на метод со специальным именем


```python
a, b = 1, 2
a + b, a.__add__(b)
```

```python
b + a, b.__add__(a) #одно и то же
```

```python
a = [7]
a[0], a.__getitem__(0) 
```

```python
b = [1, 2, 'dfs']
b.__getitem__(2), b[2] #одно и тоже
```



На практике заменять вызов операции на соответствующий метод НЕ НУЖНО. Это снижает читабельность кода. При этом вызов метода может выполняться медленнее, чем вызов соответствующей операции.

Операции, охватывающие разные типы, называются универсальными операциями. Универсальные операции в коде представлены как выражения или как встроенные функции


```python
list1 = list(range(5))
list1[-1], list1+list1, list1*0, len(list1), 5 in list1
```

```python
list2 = list(range(2))
list2[-2], list2+list1, list2*2+list1, len(list1)*list2, list1 in list2
```

Универсальные операции в Python обладают полиморфизмом, когда одинаковые операции могут выполняться над объектами различных типов

```python
'abcd'[-1], [1,2,3][-1], (1,2,3)[-1], {'a':2}['a']
```



Результат выполнения универсальной операции или метода с одинаковым именем, определенным для разных типов объектов, зависит от объекта, над которым выполняется операция или метод:

    результат выражения X + Y (вызов операции сложения) зависит от типа объектов X и Y
    в выражении X.method результат выполнения метода method зависит от типа объекта X



```python
1 + 2, '1' + '2' #пример зависимости от объекта  
```



Универсальные операции реализуются с помощью методов, которые начинаются и заканчиваются двумя знаками подчеркивания (dunder methods)


```python
print([attr for attr in dir(list1) if attr.startswith('_')])
```



Операции, специфичные для типа, представляют собой вызовы методов, которые НЕ обрамлены двумя знаками подчеркивания


```python
a = [1]

a.append(2)
print(a)

a.pop()

print(a)
```

```python
'1'.__add__('2')
```

#### Операции, специфичные для кортежей

### 2.10 Выражение спискового включения



Списковое включение (list comprehension) представляет собой способ построения НОВОГО списка за счет выполнения выражения на каждом элементе последовательности (итерируемого объекта), по одному за раз, слева направо

Целесообразно применять списковое включение тогда, когда необходимо выполнить выражение на каждом элементе последовательности (итерируемого объекта)

Выражение спискового включения записывается в квадратных скобках и содержит конструкцию цикла for ... in ...

Cписковое включение имеет следующий синтаксический шаблон:

[expr(elem) for elem in sequence]


```python
[x+'0' for x in 'string']
```

```python
[x^2+x-3 for x in range(5)]
```



Синтаксис списковых включений позволяет дополнительно задавать ограничения на элементы из последовательности (итерируемого объекта) с использованием конструкции выбора if, на которых будет выполняться выражение.

Cписковое включение с условием имеет следующий синтаксический шаблон:

[expr(elem) for elem in sequence if condition(elem)]

Списковое включение может содержать несколько конструкций выбора if, следующих друг за другом

```python
[i for i in range(100) if i > 10 if i < 30 if i % 2]
```



Синтаксис списковых включений позволяет организовать несколько уровней вложенности для прохода по нескольким последовательностям

Cписковое включение с вложенными циклами имеет следующий синтаксический шаблон:

[expr(elem_ext,elem) for elem_ext in sequence_ext for elem in sequence_int]


```python
words = ['word1', 'word2', 'word3']  
letters = [letter for word in words for letter in word]  
print(letters)
```

```python
arr = [[1,2,3], [4,5,6], [7,8,9]]
els = [el for el in arr]
els1 = [num for el in arr for num in el]
els1
```



Списковые включения могут быть вложенными друг в друга


```python
rows = 3; columns = 5
matrix = [[f'a{i}{j}' for j in range(columns)] for i in range(rows)]
matrix
```

В Python З.Х и 2.7 не только списки, но и словари, множества и генераторные объекты могут быть построены по аналогии с синтаксисом списковых включений

### 2.11 Создание строкового объекта на основе подстановок (форматирование строк)



Существует три способа создания строкового объекта на основе подстановок в исходную строку объектов произвольных типов:

    бинарная операция %: '...%s...%d...' % (obj1, obj2)
    строковый метод format: '...{}...{}...'.format(obj1, obj2)
    форматированный строковый литерал f'...{x}...{y}...'

Способ на основе бинарной операции % существует с момента создания Python. Он основан на синтаксисе функции printf языка С и часто встречается в коде Python.

Способ на основе строкового метода format появился в версии Python 2.6. Он частично совпадает по функциональности с использованием операции %.

Способ на основе форматированного строкового литерала (f-string) появился в версии Python 3.6.

Приведем пример создания строкового объекта на основе подстановок в исходную строку пяти объектов различных встроеных типов (1, 2.34, 5.6J, [7], {8})


```python
objs = (1, 2.34, 5.6J, [7], {8})
```

```python
'int: %s, float: %s, complex: %s, list: %s, set: %s' % objs
```

```python
"first %s ,second %s and the last %s" % (1, set(range(5)), [1+3j, [1-3j]])
```

### 2.12 Файловый объект

Файловый объект в Python создается вызовом встроенной функции open(file, mode), аргументами которой являются имя внешнего файла file в виде строки и режим обработки файла mode в виде строки

```python
file1 = open('tmp.txt','w');
print(file1)
```

```python
file1 = open('tmp.txt','rt');
print(file1)
```



Созданный файловый объект file предоставляет интерфейс для работы с внешним файлом на компьютере 'tmp.txt'

Python З.Х проводит четкое различие между текстовыми и двоичными данными в файлах.

    Текстовые файлы представляют содержимое в виде строк типа str, автоматически выполняя кодирование и декодирование Unicode при записывании и чтении данных.
    Двоичные файлы представляют содержимое в виде специальных строк bytes и позволяют получать доступ к содержимому файла, не изменяя его

Встроенный тип данных байтовая строка (bytes) — это последовательность символов для хранения и представления байтовой информации, используемой в файлах медиа-данных и при передаче через сеть

Некоторые режимы обработки файла:


    'r' — открыть для чтения; задается по умолчанию
    'w' — открыть для записи; если файл существует, то удалить содержимое файла
    'a' — открыть для дозаписи в конец файла; если файл существует, то содержимое файла не удалять
    'b' — открыть в двоичном режиме
    't' — открыть в текстовом режиме; задается по умолчанию
    '+' — открыть для чтения и записи

Различные режимы обработки файла могут быть объединены в одну строку. Например, режим обработки файла 'rb' означает открыть для чтения в двоичном режиме


#### Чтение файла и запись в файл

Эффективным способом построчного чтения файла в Python является использование цикла for для файлового объекта, не вызывая метод readline

```python
file = open('tmp.txt')
for line in file:
    print(line)
```



Метод readlines файлового объекта читает весь файл в список строк


```python
file = open('tmp.txt')
file.readlines()
```



Обратите внимание, что проход по файловому объекту можно сделать только один раз! Поэтому, в частности, перед вызовами методов read, readlines необходимо повторное создание файлового объект с помощью функции open.

Метод write(string) файлового объекта записывает одну строку string в файл

```python
file = open('tmp.txt','a') # открыть для дозаписи в конец файла 
file.write(str(100) + '\n')
file = open('tmp.txt')
file.read()
```



Обратите внимание, что аргументом метода write является строка, автоматического преобразования типов происходить не будет. Также символ конца строки \n автоматически добавляться не будет.

Метод writelines(list_of_strings) файлового объекта записывает список строк list_of_strings в файл


```python
file = open('tmp.txt','a') # открыть для дозаписи в конец файла
file.writelines(['200 \n', '300 \n'])
file = open('tmp.txt','r')
file.read()
```

Файловые объекты автоматически закрываются при сборке мусора. Правда, не всегда легко узнать, когда это произойдет. Оператор with для файлового объекта гарантирует его закрытие. Файл закроется автоматически после выполнения вложенного блока оператора with.

```python

with open('tmp.txt') as file:
    for line in file:
        print(line)
```
