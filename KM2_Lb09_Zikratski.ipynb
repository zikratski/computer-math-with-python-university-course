```python
class BinaryTree:
    def __init__(self):
        self.root = EmptyNode()
        BinaryNode.count = 0

    def __repr__(self):
        return repr(self.root)

    def insert(self, value):
        self.root = self.root.insert(value)

    def __contains__(self, value):
        return value in self.root
    def __len__(self):
        return len(self.root) 
    def lcr(self):
        return self.root.lcr()
```

```python
class EmptyNode:
    def __repr__(self):
        return '*'

    def insert(self, value):
        return BinaryNode(self, value, self)
    def __contains__(self, value):
        return False
    def __len__(self):
        return 0
    def lcr(self):
        return []
```

```python
class BinaryNode:
    count = 0 #создаем атрибут класса
    def __init__(self, left, value, right):
        self.left = left
        self.value = value
        self.right = right
        BinaryNode.count += 1 #при создании экземпляра, т.е. узла, увеличиваем счетчик

    def __repr__(self):
        return f'({self.left}, {self.value}, {self.right})'

    def insert(self, value):
        if value:
            if value<self.value:
                self.left = self.left.insert(value)
            elif value>=self.value:
                self.right = self.right.insert(value)
            return self
    def __contains__(self, value):
        if value or value == 0:
            if value<self.value:
                return self.left.__contains__(value)
            elif value>self.value:
                return self.right.__contains__(value)
            elif value==self.value:
                return True
    def __len__(self):
        return self.count
    def lcr(self):
        ...
        result = self.left.lcr() + [self.value] + self.right.lcr()
        ...
        return result
```

## Задание 9.1 (Центрированный обход бинарного дерева поиска)

```python
import matplotlib.pyplot as plt
import random
```

```python
source_data = [random.uniform(0,100) for i in range(10**3)]
```

```python
tree = BinaryTree()
for i in source_data:
    tree.insert(i)
```

```python
sorted_data = tree.lcr()
```

```python
plt.plot(source_data,'r.', label='Initial data') #значения точек по горизонтали, их индекс в массиве - по вертикали
plt.plot(sorted_data,' .',label='Sorted data')
plt.title('Sequences of numbers')
plt.legend()
```

## Задание 9.2 (Нахождение минимального и максимального значения)

```python
class BinaryTree:
    def __init__(self):
        self.root = EmptyNode()
        BinaryNode.count = 0

    def __repr__(self):
        return repr(self.root)

    def insert(self, value):
        self.root = self.root.insert(value)

    def __contains__(self, value):
        return value in self.root
    def __len__(self):
        return len(self.root) 
    def lcr(self):
        return self.root.lcr()
    def __min__(self):
        return self.root.__min__()
    def __max__(self):
        return self.root.__max__()
```

```python
class EmptyNode:
    def __repr__(self):
        return '*'

    def insert(self, value):
        return BinaryNode(self, value, self)
    def __contains__(self, value):
        return False
    def __len__(self):
        return 0
    def lcr(self):
        return []
    def __min__(self):
        return None
    def __max__(self):
        return None
```

```python
class BinaryNode:
    count = 0 #создаем атрибут класса
    def __init__(self, left, value, right):
        self.left = left
        self.value = value
        self.right = right
        BinaryNode.count += 1 #при создании экземпляра, т.е. узла, увеличиваем счетчик

    def __repr__(self):
        return f'({self.left}, {self.value}, {self.right})'

    def insert(self, value):
        if value:
            if value<self.value:
                self.left = self.left.insert(value)
            elif value>=self.value:
                self.right = self.right.insert(value)
            return self
    def __contains__(self, value):
        if value or value == 0:
            if value<self.value:
                return self.left.__contains__(value)
            elif value>self.value:
                return self.right.__contains__(value)
            elif value==self.value:
                return True
    def __len__(self):
        return self.count
    def lcr(self):
        ...
        result = self.left.lcr() + [self.value] + self.right.lcr()
        ...
        return result
    def __min__(self):
        if isinstance(self.left, EmptyNode):
            return self.value
        else:
            return self.left.__min__()
    def __max__(self):
        if isinstance(self.right, EmptyNode):
            return self.value
        else:
            return self.right.__max__()
            
```
